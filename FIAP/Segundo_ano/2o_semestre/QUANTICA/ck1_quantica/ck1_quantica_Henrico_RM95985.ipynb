# Imports

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_bloch_multivector, plot_histogram
from qiskit.providers.aer import AerSimulator
from qiskit import QuantumCircuit, transpile, assemble, Aer, execute
from qiskit.visualization import plot_bloch_multivector, plot_histogram
```

# Configurações dos circuitos quanticos e portas logicas

```python

# Crie um circuito quântico com um qubit
qc = QuantumCircuit(1)

# Porta X (NOT gate)
qc.x(0)

# Porta Y
qc.y(0)

# Porta Z
qc.z(0)

# Porta Hadamard (H gate)
qc.h(0)

# Porta CNOT (Controlled-X gate)
qc = QuantumCircuit(2)
qc.cx(0, 1)

simulator = Aer.get_backend('statevector_simulator')

# Execute o circuito no simulador
compiled_circuit = transpile(qc, simulator)
job = assemble(compiled_circuit)

result = execute(compiled_circuit, simulator).result()
```

---

### Explicação das Portas Lógicas

> **Porta X (NOT gate)**: Essa porta lógica inverte o estado de um qubit. Se o qubit estiver em |0⟩, ele se tornará |1⟩ e vice-versa. O operador X é muitas vezes chamado de operador de negação bit (bit-flip gate).<br>

> **Porta Y**: A porta Y é uma combinação complexa de operadores de Pauli que realiza uma rotação de 180 graus em torno do eixo Y no espaço de Bloch. Ela é como a porta X com uma rotação adicional no eixo Y.<br>

> **Porta Z**: A porta Z aplica uma fase global de -1 ao qubit. Isso não afeta diretamente os valores dos qubits em |0⟩ e |1⟩, mas afeta as superposições.<br>

> **Porta Hadamard (H gate)**: A porta H cria superposição, transformando |0⟩ em (|0⟩ + |1⟩) / √2 e |1⟩ em (|0⟩ - |1⟩) / √2. É frequentemente usado em algoritmos quânticos para criar estados superpostos.<br>

> **Porta CNOT (Controlled-X gate)**: A porta CNOT é uma porta de dois qubits. Ela funciona da seguinte forma: se o primeiro qubit (o controle) estiver em |1⟩, ela aplica uma porta X no segundo qubit (o alvo); caso contrário, não faz nada. É uma das portas lógicas mais comuns em circuitos quânticos e é fundamental para a construção de algoritmos quânticos.<br>
